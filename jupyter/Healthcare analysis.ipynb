```python
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("studentscore.csv")
```

```python
data=data.drop("Unnamed: 0" , axis=1)
data.head()
```

```python
ax = sb.countplot(data = data , x = "Gender")
ax.bar_label(ax.containers[0])
plt.show()
```

```python
gd = data.groupby("ParentEduc").agg({"MathScore" : 'mean', "ReadingScore" : 'mean', "WritingScore": 'mean'})
print(gd)
```

```python
sb.heatmap(gd, annot = True)
plt.title("Relation between Parent's Education and Student's Score")
plt.show()
```

```python
gd = data.groupby("ParentMaritalStatus").agg({"MathScore" : 'mean', "ReadingScore" : 'mean', "WritingScore": 'mean'})
print(gd)
```

```python
sb.heatmap(gd, annot = True)
plt.title("Relation between Parent's Marragestatus and Student's Score")
plt.show()
```
